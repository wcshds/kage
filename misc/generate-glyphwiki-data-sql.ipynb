```python
with open("../data/dump_newest_only.txt", encoding="utf8") as f:
    newest_data = f.readlines()
with open("../data/dump_all_versions.txt", encoding="utf8") as f:
    all_version_data = f.readlines()

data = {}
name_list = []
for line in newest_data[2:-2]:
    tmp = line.split("|")

    name = tmp[0].strip()
    related = tmp[1].strip()
    glyph_data = tmp[2].strip()

    data[name] = (related, glyph_data)
    name_list.append(name)

for line in all_version_data[2:-3]:
    tmp = line.split("|")

    name = tmp[0].strip().replace(r"\@", "@")
    related = tmp[1].strip()
    glyph_data = tmp[2].strip()

    data[name] = (related, glyph_data)
```

```python
import json


def extract_all_relevant_components(name: str) -> dict[str, (str, str)]:
    try:
        (related, glyph_data) = data[name]
    except:
        print(f"NOT FOUND: {name}")
        return {}
    result: dict[str, str] = {name: (related, glyph_data)}

    for line in glyph_data.split("$"):
        line = line.strip()
        if not line:
            continue

        fields = line.split(":")
        if not fields:
            continue

        if fields[0] == "99":
            component_name = fields[7].strip()

            if component_name and component_name not in result:
                sub_result = extract_all_relevant_components(component_name)
                result.update(sub_result)

    return result


template = f"""DROP TABLE IF EXISTS Kage;
CREATE TABLE IF NOT EXISTS Kage (GlyphID INTEGER PRIMARY KEY, GlyphName TEXT, RelatedName TEXT, GlyphData TEXT, RecursiveGlyphData TEXT);"""

with open("kage.sql", "w", encoding="utf8") as f:
    f.write(template)

values = ""
tmp = name_list[:500000]
for idx, name in enumerate(tmp):
    recursive_data = []
    for key, value in extract_all_relevant_components(name).items():
        recursive_data.append([key, value[1]])
    (related, direct_data) = data[name]
    json_recursive_data = json.dumps(recursive_data)

    values += f"""({idx}, '{name}', '{related}', '{direct_data}', '{json_recursive_data}')"""
    if idx % 25 != 0 and idx != len(tmp) - 1:
        values += ","

    if idx % 25 == 0 or idx == len(tmp) - 1:
        print(idx)
        with open("kage.sql", "a", encoding="utf8") as f:
            f.write(
                f"\nINSERT INTO Kage (GlyphID, GlyphName, RelatedName, GlyphData, RecursiveGlyphData) VALUES {values};"
            )
        values = ""
```
